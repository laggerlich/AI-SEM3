## Задание 2: 
### В папке находится изображение palm.png с пальмой. Используя классические аффинные преобразования, необходимо создать "тень" для этой пальмы и отрисовать ее с ней на одном рисунке.

```python
image2 = cv.imread("images/palm.png")
image2 = cv.cvtColor(image2, cv.COLOR_BGR2RGB)

plt.imshow(image2)
```

```python
height, width = image2.shape[:2]
width *= 2
pts1 = [
    [0, 0],
    [512, 0],
    [0, 512],
    [512, 512],
]
pts2 = [
    [600, 400],
    [1000, 400],
    [100, 512],
    [500, 512],
]

# image2_mod = image2.copy()
image2_mod = np.zeros((height, width, 3), np.uint8)
image2_mod.fill(255)
image2_mod[0:512, 0:512, :] = image2
BLACK = (0, 0, 0)
for p1, p2 in zip(pts1, pts2):
    cv.circle(image2_mod, center=p1, radius=15, color=BLACK, thickness=-1)
    cv.circle(image2_mod, center=p2, radius=15, color=BLACK, thickness=-1)
    cv.arrowedLine(image2_mod, p1, p2, color=BLACK, thickness=6) 
    
plt.imshow(image2_mod)
```

```python
M = cv.getPerspectiveTransform(np.float32(pts1), np.float32(pts2))
image2_res = cv.warpPerspective(image2, M, (width, height))

image2_res[image2_res == 0] = 255

image2_res[image2_res != 255] = 0

plt.imshow(image2_res, cmap="grey")
```

```python
image2_final = image2_res.copy()
for x in range(image2.shape[0]):
    for y in range(image2.shape[1]):
        if (image2[x,y] != 255).all():
            image2_final[x,y] = image2[x,y]

plt.imshow(image2_final, cmap="grey")
```
